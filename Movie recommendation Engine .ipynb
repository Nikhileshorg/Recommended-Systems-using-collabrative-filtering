Movie Based Recommendation Engine. Recommends similiar movies for a particular movie choosen.


```python
import numpy as np  
import pandas as pd

ratings_list = pd.read_csv("D:\\Machine learning with R\\ml-latest-small\\ratings.csv")  
ratings_list.head(2)  ##consists of 10000 + movies and ratings
```

```python
movie_titles = pd.read_csv("D:\\Machine learning with R\\ml-latest-small\\movies.csv")  
movie_titles.head(2)
```

```python
movie_list = pd.merge(ratings_list, movie_titles, on='movieId')
```

```python
movie_list.head(2) 
```

```python
movie_list.groupby('title')['rating'].mean().head(2)  
```

```python
movie_list.groupby('title')['rating'].mean().sort_values(ascending=False).head(2)  
```

```python
movie_list.groupby('title')['rating'].count().sort_values(ascending=False).head()  
```

```python
ratings_mean_count = pd.DataFrame(movie_list.groupby('title')['rating'].mean()) 
```

```python
ratings_mean_count['rating_counts'] = pd.DataFrame(movie_list.groupby('title')['rating'].count()) 
```

```python
ratings_mean_count.head(2)
```

```python
user_movie_rating = movie_list.pivot_table(index='userId', columns='title', values='rating') 
```

```python
user_movie_rating.head()
```

```python
Toy_Story_ratings = user_movie_rating['Toy Story (1995)']  
Happy_Gilmore_ratings = user_movie_rating['Happy Gilmore (1996)']
```

```python
Toy_Story_ratings.head() 
```

```python
movies_like_Toy_Story = user_movie_rating.corrwith(Toy_Story_ratings)

corr_Toy_Story = pd.DataFrame(movies_like_Toy_Story, columns=['Correlation'])  
corr_Toy_Story.dropna(inplace=True)  
corr_Toy_Story.head()  
```

```python
corr_Toy_Story.sort_values('Correlation', ascending=False).head(10) 
```

```python
corr_Toy_Story = corr_Toy_Story.join(ratings_mean_count['rating_counts'])  
corr_Toy_Story.head()
```

```python

print("----------------------------------------------------------------------")
print("////////////////Recommendation process going on/////////////////////:")
print("----------------------------------------------------------------------")
corr_Toy_Story[corr_Toy_Story ['rating_counts']>50].sort_values('Correlation', ascending=False).head()  
```
